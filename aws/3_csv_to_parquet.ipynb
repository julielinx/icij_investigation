```python
dtype_dict = {"nodes-addresses":{'int64': ['node_id'],
                                 'string': ['address',
                                            'name',
                                            'countries',
                                            'country_codes',
                                            'sourceID',
                                            'valid_until',
                                            'note']
                                },
             "nodes-officers":{'int64': ['node_id'],
                               'string': ['name',
                                          'countries',
                                          'country_codes',
                                          'sourceID',
                                          'valid_until',
                                          'note']
                              },
             "nodes-others":{'int64': ['node_id'],
                             'string': ['name',
                                        'type',
                                        'jurisdiction',
                                        'jurisdiction_description',
                                        'countries',
                                        'country_codes',
                                        'sourceID',
                                        'valid_until',
                                        'note'],
                             'datetime': ['incorporation_date',
                                          'struck_off_date',
                                          'closed_date']
                            },
             "nodes-entities":{'int64': ['node_id',
                                         'internal_id'],
                               'string': ['name',
                                          'original_name',
                                          'former_name',
                                          'jurisdiction',
                                          'jurisdiction_description',
                                          'company_type',
                                          'address',
                                          'status',
                                          'service_provider',
                                          'ibcRUC',
                                          'countries',
                                          'country_codes',
                                          'sourceID',
                                          'valid_until',
                                          'note'],
                               'datetime': ['incorporation_date',
                                            'inactivation_date',
                                            'struck_off_date',
                                            'dorm_date']
                              },
             "nodes-intermediaries":{'int64': ['node_id',
                                               'internal_id'],
                                     'string': ['name',
                                                'status',
                                                'address',
                                                'countries',
                                                'country_codes',
                                                'sourceID',
                                                'valid_until',
                                                'note']
                                    },
             "relationships":{'int64': ['node_id_start',
                                        'node_id_end'],
                              'string': ['rel_type',
                                         'link',
                                         'status',
                                         'sourceID'],
                              'datetime': ['start_date',
                                           'end_date']
                             }}
```

```python
for file in file_names:
    file_name = file.split(".")[0]
    print(file_name)
    df = pd.read_csv(os.path.join(ori_path, file))
    print(file_name + " read in. Updating dtypes")
    
    numeric_lst = dtype_dict[file_name]['int64']
    str_lst = dtype_dict[file_name]['string']
    try:
        date_lst = dtype_dict[file_name]['datetime']
    except:
        pass
    
    df[numeric_lst] = df[numeric_lst].apply(pd.to_numeric, errors='coerce')
    df[str_lst] = df[str_lst].astype('string')
    try:
        df[date_lst] = df[date_lst].apply(pd.to_datetime)
    except:
        pass
    
    print(f"Dtypes updated. Saving file {file_name}")
#     s3_path = f's3"//{s3_bucket}/{ori_path}/{file_name}.parquet'
#     print(s3_path)
#     df.to_parquet(s3_path, index=False)
    
    df_dict[file_name] = df
    print(file_name + " in dict")
    new_path = os.path.join(os.getcwd(), 'data', 'compressed', file_name)
    df.to_parquet(f"{new_path}.parquet", index=False)
```
